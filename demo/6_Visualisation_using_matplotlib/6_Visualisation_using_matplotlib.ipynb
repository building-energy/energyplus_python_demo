## Plot the external air temperature

```python
ext_temp=df['Environment:Site Outdoor Air Drybulb Temperature [C](Hourly)']
ext_temp
```

'ext_temp' is a pandas Series instance and can be plotted using the plot method...

```python
ext_temp.plot()
```

## To improve this plot...

### Read the Date/Time column as pandas TimeStamps

```python
def parse_datetime_string(st):
    st=st.strip()
    year=2001
    month=int(st[0:2])
    day=int(st[3:5])
    hour=int(st[7:9])
    minute=int(st[10:12])
    second=(st[13:15])
    if not hour==24:
        dt=pd.Timestamp(year,month,day,hour,minute)
    else:
        hour=0
        dt=pd.Timestamp(year,month,day,hour,minute)
        dt+=pd.Timedelta('1 day')
    return dt
l=df['Date/Time']
l=[parse_datetime_string(i) for i in l]
datetime=pd.Series(data=l).values
datetime
```

### then plot using matplotlib directly

```python
fig=plt.figure(figsize=(14,4))
ax=fig.add_axes([0, 0, 0.5, 1])
ax.set_xlabel('Date')
ax.set_ylabel('Hourly external air temperature (C)')
ax.plot(datetime,ext_temp.values,'-')
plt.show()
print('Figure X: Hourly external air temperature for the simulation period')
```

### save the figure

```python
fig.savefig('Figure_X.png',bbox_inches='tight', pad_inches=0.25)
```

## Plot results from different simulation runs

This plots the zone air temperature for the 4 orientations for July...

```python
fig=plt.figure(figsize=(14,4))
ax=fig.add_axes([0, 0, 0.5, 1])
ax.set_xlabel('Date')
ax.set_ylabel('Hourly zone mean air temperature (C)')
orientations=[0,90,180,270]
for orientation in orientations:
    fp='detached_house_{}out.csv'.format(orientation)
    df=pd.read_csv(fp)
    index=pd.Series(data=[parse_datetime_string(i) for i in df['Date/Time']]).values
    data=df['LIVING_ROOM:Zone Mean Air Temperature [C](Hourly)'].values
    s=pd.Series(index=index,
                data=data)
    mask=((s.index>=pd.Timestamp(2001,7,1)) & (s.index<pd.Timestamp(2001,8,1)))
    s=s[mask]
    ax.plot(s.index,s.values,'-',label=str(orientation))
ax.legend()
plt.show()
print('Figure Y: Impact of orientation on the hourly zone mean air temperature.')
```

### save the figure

```python
fig.savefig('Figure_Y.png',bbox_inches='tight', pad_inches=0.25)
```
